```python
import requests
from bs4 import BeautifulSoup
from datetime import date
import datetime
import csv

# Get the current date
today = datetime.datetime(2024, 1, 21).date()
#today = date.today()

# Store the dataset
filename = '/content/drive/MyDrive/1 - PhD Datasets for Summarization/Yumsuk-HNLP-bbchausa.csv'


# Function to extract article information
def extract_article_info(article_url):
    response = requests.get(article_url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'lxml')
        headline_elem = soup.find('div', class_='bbc-1151pbn ebmt73l0')
        article_date_elem = soup.find('div', class_='bbc-19j92fr ebmt73l0').find('time', class_='bbc-j3wi5n e1mklfmt0')
        article_content_elem = soup.find('main', class_='bbc-fa0wmp')

        if headline_elem and article_date_elem and article_content_elem:
            headline = headline_elem.text.strip()
            article_date = article_date_elem.get('datetime')[:10]  # Extract only the date part
            article_content = ''
            summary = ''
            results = soup.find_all('div', class_='bbc-19j92fr ebmt73l0')

            if len(results) > 1:
                summary = str(results[1].text)

            for i in range(2, len(results)):
                article_content += str(results[i].text)

            return {
                'headline': headline,
                'article_date': article_date,
                'summary': summary,
                'article_content': article_content
            }

    # Return None if status_code is not 200
    return None

# Main scraping function
def scrape_bbc_hausa():
    url = 'https://www.bbc.com/hausa'
    response = requests.get(url)

    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'lxml')
        articles = soup.find_all('div', class_='promo-text')

        with open(filename, 'a', newline='', encoding='utf-8') as file:
            csvwriter = csv.writer(file)
            #csvwriter.writerow(['Date', 'HeadLines', 'Summary', 'Articles'])
            for article in articles:
                article_url = article.find('a', class_='focusIndicatorDisplayBlock bbc-uk8dsi e1d658bg0').get('href')

                if article_url.startswith('https://www.bbc.com/hausa/articles/'):
                    data_set = extract_article_info(article_url)

                    if data_set and data_set['article_date'] == today.isoformat():
                        if data_set['article_content']:
                            #csvwriter.writerow([data_set['article_date'],
                             #                   data_set['headline'],
                             #                    data_set['summary'],
                              #                  data_set['article_content']])
                            print(data_set['article_date'])
                            print(data_set['headline'])
                            print('Summary')
                            print(data_set['summary'])
                            print('Article')
                            print(data_set['article_content'])
                            print()
                            #print('Saved Successfully')

# Run the scraping function
scrape_bbc_hausa()
```

```python
import requests
from bs4 import BeautifulSoup
from datetime import datetime
import csv

def get_page_content(url):
    response = requests.get(url)
    return BeautifulSoup(response.content, 'html.parser')

def get_today_date():
    return datetime.now().strftime('%Y-%m-%d')

def scrape_and_save_data(url, csv_filename):
    soup = get_page_content(url)
    title = soup.find('title').text
    print(title)
    contents = soup.find_all('div', class_='media-block__content')
    print()

    with open(csv_filename, 'a', newline='', encoding='utf-8') as file:
        csvwriter = csv.writer(file)
        #csvwriter.writerow([pub_date, article_content, headline])

        for content in contents:
            date_span = content.find('span')

            if date_span:
                headline = content.find('h4').text.strip()
                article_url = content.find('a').get('href').strip()
                new_url = f'https://www.voahausa.com/{article_url}'

                inner_soup = get_page_content(new_url)
                date_time = inner_soup.find('time')

                if date_time:
                    pub_date = date_time.get('datetime')[:10]
                    today_date = get_today_date()
                    #print(today_date)   2024-01-21

                    if pub_date == '2024-01-19':
                        news = inner_soup.find('div', class_='content-floated-wrap fb-quotable')
                        if news:
                            paragraphs = news.find_all('p')[:-3]
                            article_content = ' '.join(paragraph.text for paragraph in paragraphs)
                            #csvwriter.writerow([pub_date,
                             #                  article_content,
                             #                headline])
                             #print('Saved Successfully')


                            print('Saved Successfully')

                            print(f'Headline: {headline}')
                            print(f'Date: {pub_date}')
                            print('Article:')
                            print(article_content)
                            print()

# Specify the URL and CSV filename
url_to_scrape = 'https://www.voahausa.com/'
csv_filename = '/content/drive/MyDrive/1-PhD Thesis Dataset/Yumsuk-HNLP-voahausa.csv'

# Call the function to scrape and save data
scrape_and_save_data(url_to_scrape, csv_filename)
```
